```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("smartcart_customers.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

## Handling Missing Values

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

## Feature Engineering

```python
# Age
df["age"] = 2026 - df["Year_Birth"]
```

```python
# Customer Joining Date
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)

reference_date = df["Dt_Customer"].max()

df["Customer_Tenure_Days"] = (reference_date - df["Dt_Customer"]).dt.days
```

```python
# Spending
df["Total_Spending"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
```

```python
# Childrens
df["Children"] = df["Kidhome"] + df["Teenhome"]
```

```python
df["Education"].value_counts()
```

```python
# Education --> UnderGrad, Grad, PostGrad

df["Education"] = df["Education"].replace({
    "Basic": "UnderGraduate", "2n Cycle": "UnderGraduate",
    "Graduation": "Graduate",
    "PhD": "PostGraduate", "Master": "PostGraduate"
})
```

```python
df["Marital_Status"].value_counts()
```

```python
# Marital Status

df["Living_With"] = df["Marital_Status"].replace({
    "Married": "Partner", "Together": "Partner",
    "Single": "Alone", "Divorced": "Alone", "Widow": "Alone",
    "Alone": "Alone", "Absurd": "Alone", "YOLO": "Alone"
})
```

```python
df.head()
```

## Drop Cols

```python
cols = ["ID", "Year_Birth", "Marital_Status", "Kidhome", "Teenhome", "Dt_Customer"]
spending_cols = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
```

```python
cols_to_drop = cols + spending_cols

df_cleaned = df.drop(columns=cols_to_drop)
```

```python
df_cleaned.columns
```

## Outliers

```python
cols = ["Recency", "Income", "age", "Total_Spending", "Children"]

# Visualizing with pairplot
plt.figure(figsize=(12, 12))
sns.pairplot(df_cleaned[cols])
plt.tight_layout()
```

```python
# Remove Outliers
print(f"Data size with outliers: {len(df_cleaned)}")

df_cleaned2 = df_cleaned[ (df_cleaned["age"] < 90) ]
df_cleaned2 = df_cleaned2[ (df_cleaned2["Income"] < 600_000) ]

print(f"Data size without outliers: {len(df_cleaned2)}")
```

```python
# Corr Heatmap

corr = df_cleaned2.corr(numeric_only=True)
```

```python
plt.figure(figsize=(9, 8))

sns.heatmap(
    data=corr,
    annot=True,
    annot_kws={"size": 6},
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.tight_layout()
```

## Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()  # No drop_first=True beacuse we don't need to do it in case of clustering as it migth derive some useful info and do not have any problem with having multicolinearity

cat_cols = ["Education", "Living_With"]

enc_cols = ohe.fit_transform(df_cleaned2[cat_cols])
```

```python
enc_df = pd.DataFrame(enc_cols.toarray(), columns=ohe.get_feature_names_out(), index=df_cleaned2.index)
```

```python
enc_df.head()
```

```python
df_encoded = pd.concat([df_cleaned2.drop(columns=cat_cols), enc_df], axis=1)
```

```python
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df_encoded
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Visualizing

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2])

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D Projection")
```

## Analyze value of K

### 1) Elbow Method

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
```

```python
knee = KneeLocator(range(1, 11), wcss, curve="convex", direction="decreasing")
optimal_k = knee.elbow
```

```python
print(f"Optimal k = {optimal_k}")
```

```python
# plot

sns.lineplot(x=range(1, 11), y=wcss, marker="o")

plt.xlabel("K value")
plt.ylabel("wcss")
plt.title("knee plot")
```

### 2) Silhouette Score

```python
from sklearn.metrics import silhouette_score

scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, labels)
    scores.append(score)
```

```python
# plot

plt.plot(range(2, 11), scores, marker="o")

plt.xlabel("K value")
plt.ylabel("silhouette_score")

plt.title("finding optimal k")
```

```python
# combined plot

# We take the value of k which is nearest to the point where silhouette score and wcss plot intersect

fig, ax1 = plt.subplots(figsize=(8, 6))

ax1.plot(range(2, 11), wcss[1: 11], marker="o", color="blue")
ax1.set_xlabel("K")
ax1.set_ylabel("wcss")

ax2 = ax1.twinx()
ax2.plot(range(2, 11), scores, marker="o", color="red", linestyle="--")
ax2.set_ylabel("silhoette_score")
```

```python
kmeans = KMeans(n_clusters=4, random_state=22)
kmeans_labels = kmeans.fit_predict(X_pca)

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=kmeans_labels)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D Projection")
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=4, linkage="ward")
agg_labels = agg.fit_predict(X_pca)

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=agg_labels)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D Projection")
```

## Characterization of Clusters

```python
df_cleaned2["clusters"] = agg_labels
```

```python
df_cleaned2.head()
```

```python
pal = ["red", "blue", "green", "yellow"]
sns.countplot(x=df_cleaned2["clusters"], palette=pal, hue=df_cleaned2["clusters"])
```

```python
# Income and Spending Patterns

sns.scatterplot(x=df_cleaned2["Income"], y=df_cleaned2["Total_Spending"], hue=df_cleaned2["clusters"], palette=pal)
```
